```python
import numpy as np
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,2))
df2=scaler.fit_transform(np.array(df2).reshape(-1,1))
```

```python
print(df2)
```

```python
##split dataset into train
trainee_size=int(len(df2)*0.68)
test_size=len(df2)-trainee_size
train_data,test_data=df2[0:trainee_size,:],df2[trainee_size:len(df2),:1]
```

```python
trainee_size,test_size
```

```python
train_data
```

```python
import numpy
# convert an array of values into a dataset matrix
def create_dataset(dataset, time_step=1):
  dataX, dataY = [], []
  for i in range(len(dataset)-time_step-1):
    a = dataset[i:(i+time_step), 0]
    dataX.append(a)
    dataY.append(dataset[i + time_step, 0])
  return numpy.array(dataX), numpy.array(dataY)
```

```python
time_step = 100
b_train, v_train = create_dataset(train_data, time_step)
b_test, vtest = create_dataset(test_data, time_step)
```

```python
print(b_train.shape), print(v_train.shape)
```

```python
print(b_test.shape), print(vtest.shape)
```

```python
b_train =b_train.reshape(b_train.shape[0],b_train.shape[1] , 1)
b_test = b_test.reshape(b_test.shape[0],b_test.shape[1] , 1)
```

```python
##Create thw LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(b_train,v_train,validation_data=(b_test,vtest),epochs=100,batch_size=64,verbose=1)
```

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
##predict and check the performances
train_predict=model.predict(b_train)
test_predict=model.predict(b_test)
```

```python
##transforming to original form'train_predict=scaler.inverse_transform(train_predict)
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```

```python
##Calculating the RMSE performance
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(v_train,train_predict))
```

```python
##Testing data RMSE
math.sqrt(mean_squared_error(vtest,test_predict))
```

```python
### Plot the graph
# shift train predict for plotting
look_back=100
trainPredictPlot = numpy.empty_like(df2)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(train_predict)+look_back, :] = train_predict
# shift test predict for plotting
testPredictPlot = numpy.empty_like(df2)
testPredictPlot[:, :] = numpy.nan
testPredictPlot[len(train_predict)+(look_back*2)+1:len(df2)-1, :] = test_predict
# plot baseline and predict
plt.plot(scaler.inverse_transform(df2))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```
